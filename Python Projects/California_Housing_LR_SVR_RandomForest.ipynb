# California Housing — Linear Regression vs SVR vs Random Forest

Uses the California Housing dataset to compare Linear Regression, SVR (linear kernel),
and Random Forest Regressor on house value prediction. Reports R² scores and feature
importances, and discusses overfitting and non-linearity trade-offs between models.


```python
#a. Load and Describe the Dataset

from sklearn.datasets import fetch_california_housing
import pandas as pd

# Load California Housing dataset
data = fetch_california_housing(as_frame=True)
df = data.frame

# Extract the first 1000 samples
df = df.iloc[:1000].copy()

# Create 'RoomsPerPerson' as a proxy (since 'Households' is not available)
df['RoomsPerPerson'] = df['AveRooms'] / df['AveOccup']

# Select three key features
selected_features = df[['MedInc', 'HouseAge', 'RoomsPerPerson']]
target = df['MedHouseVal']

# Display basic statistics
stats = selected_features.copy()
stats['Target'] = target
print("Basic Statistics")
stats.describe()
```

```python
#b. Train a Linear Regression Model
from sklearn.linear_model import LinearRegression

# Train the model
lr = LinearRegression()
lr.fit(selected_features, target)

# Output coefficients and intercept
print("\n Linear Regression.")
print("Coefficients:", dict(zip(selected_features.columns, lr.coef_)))
print("Intercept:", lr.intercept_)

# R² score
r2_lr = lr.score(selected_features, target)
print("R² Score (Linear Regression):", r2_lr)
```

```python
#c. Implement Support Vector Regression (SVR) with Linear Kernel
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# SVR with linear kernel and standardization
svr_model = make_pipeline(StandardScaler(), SVR(kernel='linear'))
svr_model.fit(selected_features, target)

# R² score
r2_svr = svr_model.score(selected_features, target)

print("\n Support Vector Regression (Linear Kernel)")
print("R² Score (SVR):", r2_svr)

# SVR vs Linear Regression:
print("\nNote: SVR is more robust to outliers compared to Linear Regression, as it ignores errors within a certain margin (epsilon).")
```

```python
# Train a Random Forest Regressor and Analyze Feature Importances
from sklearn.ensemble import RandomForestRegressor

# Train Random Forest Regressor
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(selected_features, target)

# R² score
r2_rf = rf.score(selected_features, target)

# Feature importance
feature_importances = dict(zip(selected_features.columns, rf.feature_importances_))

print("\nRandom Forest Regressor")
print("R² Score (Random Forest):", r2_rf)
print("Feature Importances:", feature_importances)

# Overfitting discussion
print("\nNote: Random Forests can capture complex patterns but are prone to overfitting if not properly regularized.")
```

```python
#e. Advantages of Random Forest Over SVR for Non-Linear Data
print("\nAdvantage of Random Forest over SVR")
print("Random Forest Regression handles non-linear relationships inherently by averaging many decision trees.")
print("This makes it more effective than SVR (with a linear kernel) for complex, irregular pricing patterns in real estate markets.")
```
